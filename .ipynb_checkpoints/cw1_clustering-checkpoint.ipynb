### Part 1 of the project...


K-clustering has partitions represented by <b>K</b>, where, if n is the number of inputs and K is the number of clusters, <b>K</b> is a K.n binary matrix. Below is the K-Means algorithm pseudo-code rewritten from the lecture notes. As the project specification asks for a comparison betweem different implementations of k-means with varying methods of calculating the distance,

```
Input: Data X = {x1...xn} (set of points to be clustered), Order K (how many clusters in partition), MAX (number of iterations allowed)
Output = Partition

POut = []
Randomly initialize the representatives (centroids typically) for each cluster.

for t in range(1, MAX+1):
    ASSIGNMENT:
        Assign each sample within the dataset X to the cluster with the nearest representative. The
        measurement, represented by d, can be anything. For this coursework we are using the
        Mahalanobis distance.
        
    UPDATE:
        Update the representatives of each cluster, calculating them from the samples just assigned to
        them.
        
        Update the partition with the new clusters:
            Pt = {binary matrix for different clusters}
    
    TERMINATION:
        If partition isn't changing then output it as the true partition as none of the sample are
        changing cluster when being assigned to the nearest one.

```

Covariance Matrix, S = 1/(N-1)<b>X<sup>T</sup>X</b> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;  /*nb*, when N = 1, let S = <b>X<sup>T</sup>X</b>


#### Miscellaneous Functions:

Functions that are used in the project and required to run it, but not explicitly part of the kMeans algorithm.

````python
### Takes a standard (n.k) binary matrix partition and converts it to a n.1 matrix ###
def flatten(partition):
    out = zeros(partition.shape[1])
    for i in range(1,partition.shape[0]):
        out += i*partition[i]
    return out
````

````python
### Takes a (n.1) matrix and converts it to a standard n.k partition matrix ###
def expand(flat):
    out = zeros((max(flat) + 1,flat.shape[0]))
    for i in range(flat.shape[0]):
        out[flat[i]][i] = 1
    return out
````

````python
### A way of comparing the result to the targets ###
##    Doesn't actually work as intended for when the clusters are very incorrect but is useful for comparing
##    two or more nearly correct partitions
def errors(partition, target):
    target = expand(target)
    
    partition.sort()
    target.sort()
    
    partition = flatten(partition)
    target = flatten(target)
    
    e = 0
    print(target.shape[0])
    
    for i in range(target.shape[0]):
        if target[i] != partition[i]:
            e += 1
    
    return e    
````

#### Variant Distance Functions:
A collection of different functions to measure the distance/dissimilarity between a point and the representative within the kMeans algorithm. While testing the various functions to see if they worked, I wrote them seperate from the main function to allow for greater control. Switching between Euclidean and Mahalanobis without having to change anything was immensely useful.

- **Euclidean Distance**

The simplest method of measuring the distance from a point to another. Used in testing the KMeans function before adding other methods of measuring distance, however is left for comparison.

````python
### EUCLIDEAN DISTANCE ###
def d(point, centroid, data, partition):
    out = 0
    for i in range(point.shape[0]):
        out += (point[i] - centroid[i])**2
    return sqrt(out)


### EUCLIDEAN DISTANCE INITIAL CLUSTER. FUNCTION REQUIRED TO BE SET AS MAY CHANGE WITH THE MAHALANOBIS FUNCTIONS ###
def initialCluster(data, reps, K):
    print("EUCLIDEAN DISTANCE")
    
    n = data.shape[0]
    partition = zeros((K,n))
    for i in range(n):        
        point = data[i]
        distances = zeros(K)
        for j in range(K):
            out = 0
            for k in range(point.shape[0]):
                out += (point[k] - reps[j][k])**2
            distances[j] = sqrt(out)
        partition[argmin(array(distances))][i] = 1
        
    for i in range(K):
            t = partition[i] == 1
            repData = data[t]
            reps[i] = mean(repData, axis=0,dtype=np.float32)
    return partition
````

- **Mahalanobis Distance**

As required by the coursework definition, the Mahalanobis function. As the Mahalanobis Distance uses a set of data to form the covariance, there are two different datasets that could be used. The first is all the data from all the clusters, and the second is just the data from the cluster being updated. As the calculation of the distance requires a set of data, on the initial pass, if only data from the cluster is being used, it will not work. For this reason, the "initial cluster" must be calculated using euclidean distance or using all the data. As is was unsure which would be the "best" method of implementing Mahalanobis below there is a series of variant methods based on Mahalanobis distance:
  * Using all the data, with Euclidean initial
  * Using all the data, with Mahalanobis initial
  * Using just the cluster data, with Euclidean initial
  * Using just the cluster data, with Mahalanobis initial
  
##### All data, Euclidean initial evaluation:

````python
### MAHALANOBIS DISTANCE: ALL DATA, EUCLIDEAN INITIAL ###
def d(point, centroid, data, partition):
    delta = point - centroid
    return sqrt( dot( dot( delta.T , linalg.inv(cov(data.T))), delta))

def initialCluster(data, reps, K):
    print("MAHALANOBIS DISTANCE: ALL DATA, EUCLIDEAN INITIAL") ## FOR IDENTIFYING THE SETTINGS WHEN RUNNING
    
    n = data.shape[0]
    partition = zeros((K,n))
    for i in range(n):        
        point = data[i]
        distances = zeros(K)
        for j in range(K):
            out = 0
            for k in range(point.shape[0]):
                out += (point[k] - reps[j][k])**2
            distances[j] = sqrt(out)
        partition[argmin(array(distances))][i] = 1
        
    for i in range(K):
            t = partition[i] == 1
            repData = data[t]
            reps[i] = mean(repData, axis=0,dtype=np.float32)
    return partition
````

- **Random**

The first and simpilest method to chose initial representatives. I found this method inconsistent as it can chose values all within one cluster, produceing odd results where one or more clusters only have one point assigned to it.

````python
### RANDOM INITIAL REPRESENTATIVES ###
def initialReps(data, k):
    print("RANDOM INITIAL REPRESENTATIVES")
    index = random.choice(data.shape[0], k, replace=False)
    return data[index]
````

- **Around a Circle**

As the data is always standarized, this function pick points based on evenly space points around a circle, radius 1. Works far more consistently as it avoids the issue of choosing several points too close within one cluster. Also helps compare the two distance functions by always starting with the same points. However it loses the randomness of the intial clusterings meaning the result of the kMeans algorithm is always the same.

````python
### DEFINED POINTS AROUND A CIRCLE
def initialReps(data, k):
    print("POINTS AROUND A CIRCLE INITIAL REPRESENTATIVES")
    d = data.shape[1]

    out = zeros((k,d))
    
    for i in range(k):        
        it = (2*pi*i)/k
        for j in range(d - 1):
            out[i,j] = sin(it)**j * cos(it)
        out[i,d-1] = sin(it)**(d-1)
    return out
````

- **K-means++**
An algorithm suggested in 2007 by David Arthur and Sergei Vassilvitskii, it is a way to avoid the sometimes poor initial clustering found by the other two methods and aims to seed the intial clusters in a particular way. The algorithms follows:
    1. Choose one center with a uniform randomness from among all data points.
    2. For each data point x, compute D(x), the distnace between x and the nearest center that has already been chosen
    3. Choose one new data point at random as a new center, using a weighted probability proportional to D(x)^2
    4. Repeat steps 2 and 3 until k centers have been chosen
    
This method takes far longer than the previous, however this seeding yields considerable improvement in the final error of k-means. However I found in testing that if the first, random point, was on the extremes of the dataset it would cause the same issues as just having random data, so I designed a variant that always uses the most central data point as the first seeded representative and then randomly chooses from that. This showed far more consistent results.

````python
### K-MEANS++: RANDOM FIRST POINT ###
import random

def dist(point, centroid):
    out = 0
    for i in range(point.shape[0]):
        out += (point[i] - centroid[i])**2
    return sqrt(out)

def initialReps(data, k):
    print("K-MEANS++ INITIAL REPRESENTATIVES: RANDOM PRIMARY")
    dim = data.shape[1]
    n = data.shape[0]
    
    out = zeros((k,dim))
    
    out[0] = random.choice(data)
    part = ones(n)
    
    for i in range(1,k):
        dMatrix = full(n, inf)
        for j in range(n):
            for k in range(i):
                dMatrix[j] = min(dMatrix[j], dist(data[j], out[k])**4)
        total = sum(dMatrix)
        r = random.uniform(0,total)
        for point in range(n):
            r -= dMatrix[point]
            if r <= 0:
                out[i] = data[point]
                break        
        
    return out      
        
````

````python
### K-MEANS++: CENTRAL FIRST POINT ###
import random

def dist(point, centroid):
    out = 0
    for i in range(point.shape[0]):
        out += (point[i] - centroid[i])**2
    return sqrt(out)

def initialReps(data, k):
    print("K-MEANS++ INITIAL REPRESENTATIVES: CENTER PRIMARY")
    dim = data.shape[1]
    n = data.shape[0]
    
    out = zeros((k,dim))
    
    Xs = data[:,0] ** 2
    for i in range(1,dim):
        Xs = data[:,i] ** 2
    
    out[0] = data[Xs.argsort()][0] ## SET THE FIRST SEEDED REPRESENTATIVE TO THE POINT CLOSEST TO
    part = ones(n) ## THE CENTER (0,0). STILL RANDOM, BUT LESS RANDOM AND HOPEFULLY MORE CONSISTENT
    
    for i in range(1,k):
        dMatrix = full(n, inf)
        for j in range(n):
            for k in range(i):
                dMatrix[j] = min(dMatrix[j], dist(data[j], out[k])**4)
        total = sum(dMatrix)
        r = random.uniform(0,total)
        for point in range(n):
            r -= dMatrix[point]
            if r <= 0:
                out[i] = data[point]
                break        
        
    return out      
        
````

````python
### kMEANS GENERIC MAIN FUNCTION ###
iniReps = []
def kmeans(data: array, K: int, MAX: int):
    n = data.shape[0]

    reps = initialReps(data, K)  ## INITIAL REPRESENTATIVES CHOSEN THROUGH CHOSEN METHOD
    iniReps = asarray(reps.copy())
    
    Partition = initialCluster(data, reps, K)   ## SEPERATE FOR EXPERIMENTAL PURPOSES
    
    for t in range(MAX):
        
        tempPartition = zeros((K,n))
        
        ### ASSIGNMENT PHASE ###
        for i in range(n):        
            point = data[i]
            distances = zeros(K)
            for j in range(K):
                distances[j] = d(point,reps[j],data,Partition[j]) ## DISTANCE CALCULATED THROUGH CHOSEN METHOD
            tempPartition[argmin(array(distances))][i] = 1
            
        ### UPDATE PHASE ###
        for i in range(K):
            t = tempPartition[i] == 1
            repData = data[t]
            reps[i] = mean(repData, axis=0,dtype=np.float32)
            
        ### TERMINATION PHASE ###
        if array_equal(Partition, tempPartition):
            return Partition, iniReps
        Partition = tempPartition.copy()        
    return Partition, iniReps



out, iniReps = kmeans(X,3,100)

outFlat = flatten(out)

### INFORMATION ON SUCCESS RATE ###
print("DIMENSIONS:", X.shape[1])
print("CLUSTERS:  ", max(y))


### GRAPHING RESULTS ###
plt.scatter(X[:,0],X[:,1], c=outFlat, s=150, cmap='inferno');
plt.scatter(iniReps[:,0],iniReps[:,1],s=500,marker="x")
plt.show()
from mpl_toolkits.mplot3d import Axes3D
if X.shape[1] >= 3:
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')    
    ax.scatter(X[:,0], X[:,1], X[:,2], c=outFlat, s= 100, cmap='inferno')   
    ax.scatter(iniReps[:,0],iniReps[:,1],iniReps[:,2], s= 500, marker = "x",depthshade = False)
    plt.show()
````

#### Data Generation Functions
For the k-Means algorithm to function properly (at least with Mahalanobis distance) standardized data needs to be generated. The type of data generated can vary but all of the functions are written such that they. Below are different methods for generating sets of data.

  - **Standard Blobs**
Mimicing the method used in Workshop 3 for generating reusable data, formatted into a function to allow for different number of dimensions and number of clusters easily. 

````python
### MAKE STANDARDIZED N DIMENSIONAL RESUABLE DATA ###

dim = 2 ## NUMBER OF DIMENSIONS
clus = 3 ## NUMBER OF CLUSTERS
sampleNum = 500 ## NUMBER OF SAMPLES

def generateData(dim, clus, sampleNum = 500):
    from sklearn.datasets.samples_generator import make_blobs
    X, y = make_blobs(n_samples=sampleNum,
                  n_features=dim,
                  centers=clus,
                  cluster_std=1.5,
                  center_box=(-10.0, 10.0),
                  shuffle=True)


    ### STANDARDIZE nTH DIMENSIONAL DATA ###
    for i in range(X.shape[1]):
        X[:,i] = (X[:,i]-np.mean(X[:,i]))/np.std(X[:,i])

    
    return X, y

X, y = generateData(dim, clus, sampleNum)

print(X.shape)
# plot the data and apply colours according to the correct labeling
plt.scatter(X[:, 0], X[:, 1], c=y, s=200, cmap='inferno');
plt.show()
from mpl_toolkits.mplot3d import Axes3D
if X.shape[1] >= 3:
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X[:,0], X[:,1], X[:,2], c=y, s= 300, cmap='inferno')
    plt.show()
````

   - **Stretched Blobs**
   Generates blobs of data, similar to the previous data generation function, however stretches them in one axis. It is for this data, that Mahalanobis shows significant improvement over the Euclidean methods, as the distance between the blobs is shorter than the distances within the blob.

````python
### MAKE STRETCHED STANDARDIZED N DIMENSIONAL RESUABLE DATA ###

dim = 3 ## NUMBER OF DIMENSIONS
clus = 3 ## NUMBER OF CLUSTERS
sampleNum = 1000 ## NUMBER OF SAMPLES

def generateData(dim, clus, sampleNum = 500):
    from sklearn.datasets.samples_generator import make_blobs
    X, y = make_blobs(n_samples=sampleNum,
                  n_features=dim,
                  centers=clus,
                  cluster_std=1.5,
                  center_box=(-10.0, 10.0),
                  shuffle=True)        
    
    ### ELONGATE DATA IN ONE AXIS, CONDENSE IN OTHERS (PER CLUSTER) ###
    for c in range(clus):
        Xc = X[y == c]
        Xc[:, 0] = Xc[:, 0] + 10*(Xc[:, 0] - np.mean(Xc[:, 0]))
        X[y == c] = Xc

    for i in range(1,X.shape[1]):
        for c in range(clus):
            Xc = X[y == c]
            Xc[:, i] = Xc[:, i] - 0.5*(Xc[:, i] - np.mean(Xc[:, i]))
            X[y == c] = Xc
     
    ### STANDARDIZE nTH DIMENSIONAL DATA ###
    for i in range(X.shape[1]):
        X[:,i] = (X[:,i]-np.mean(X[:,i]))/np.std(X[:,i])
        
    return X, y

X, y = generateData(dim, clus, sampleNum)

print(X.shape)
# plot the data and apply colours according to the correct labeling
figsize(15,15)
plt.scatter(X[:, 0], X[:, 1], c=y, s=150, cmap='inferno');
plt.show()
from mpl_toolkits.mplot3d import Axes3D
if X.shape[1] >= 3:
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X[:,0], X[:,1], X[:,2], c=y, s= 100, cmap='inferno')
    plt.show()
````

````python
### EUCLIDEAN DISTANCE ###
def d(point, centroid, data, partition):
    out = 0
    for i in range(point.shape[0]):
        out += (point[i] - centroid[i])**2
    return sqrt(out)


### EUCLIDEAN DISTANCE INITIAL CLUSTER. FUNCTION REQUIRED TO BE SET AS MAY CHANGE WITH THE MAHALANOBIS FUNCTIONS ###
def initialCluster(data, reps, K):   
    n = data.shape[0]
    partition = zeros((K,n))
    for i in range(n):        
        point = data[i]
        distances = zeros(K)
        for j in range(K):
            out = 0
            for k in range(point.shape[0]):
                out += (point[k] - reps[j][k])**2
            distances[j] = sqrt(out)
        partition[argmin(array(distances))][i] = 1
        
    for i in range(K):
            t = partition[i] == 1
            repData = data[t]
            reps[i] = mean(repData, axis=0,dtype=np.float32)
    return partition

def kmeans(data: array, K: int, MAX: int):
    n = data.shape[0]

    reps = initialReps(data, K)  ## INITIAL REPRESENTATIVES CHOSEN THROUGH CHOSEN METHOD
    iniReps = asarray(reps.copy())
    
    Partition = initialCluster(data, reps, K)   ## SEPERATE FOR EXPERIMENTAL PURPOSES
    
    for t in range(MAX):
        
        tempPartition = zeros((K,n))
        
        ### ASSIGNMENT PHASE ###
        for i in range(n):        
            point = data[i]
            distances = zeros(K)
            for j in range(K):
                distances[j] = d(point,reps[j],data,Partition[j]) ## DISTANCE CALCULATED THROUGH CHOSEN METHOD
            tempPartition[argmin(array(distances))][i] = 1
            
        ### UPDATE PHASE ###
        for i in range(K):
            t = tempPartition[i] == 1
            repData = data[t]
            reps[i] = mean(repData, axis=0,dtype=np.float32)
            
        ### TERMINATION PHASE ###
        if array_equal(Partition, tempPartition):
            return Partition, reps
        Partition = tempPartition.copy()        
    return Partition, reps


def dist(point, centroid):
    out = 0
    for i in range(point.shape[0]):
        out += (point[i] - centroid[i])**2
    return sqrt(out)

def initialReps(data, k):
    import random
    dim = data.shape[1]
    n = data.shape[0]
    
    out = zeros((k,dim))
    
    Xs = data[:,0] ** 2
    for i in range(1,dim):
        Xs = data[:,i] ** 2
        
    out[0] = data[Xs.argsort()][0] ## SET THE FIRST SEEDED REPRESENTATIVE TO THE POINT CLOSEST TO
    part = ones(n) ## THE CENTER (0,0). STILL RANDOM, BUT LESS RANDOM AND HOPEFULLY MORE CONSISTENT
    
    for i in range(1,k):
        dMatrix = full(n, inf)
        for j in range(n):
            for k in range(i):
                dMatrix[j] = min(dMatrix[j], dist(data[j], out[k])**4)
        total = sum(dMatrix)
        r = random.uniform(0,total)
        for point in range(n):
            r -= dMatrix[point]
            if r <= 0:
                out[i] = data[point]
                break        
        
    return out    


####                        ####
##    DAVIES-BOULDIN INDEX    ##
####                        ####

def s(i, X, partition, reps):
    clus = X[partition[i] == 1]
    normClus = len(clus)
    sumClus = 0
    for point in clus:
        sumClus += dist(point, reps[i])
    return sum(clus)/len(clus)

def Rij(i, j, n, X, partition, reps):
    return(s(i, X, partition, reps) + s(j, X, partition, reps))/dist(reps[i],reps[j])    


def R(i, n, X, partition, reps):
    listR = []
    for j in range(n):
        if(i!=j):
            listR.append(Rij(i, j, n, X, partition, reps))
    return max(listR)

def DB(n, X, partition, reps): ### CALCULATES DAVIES BOULDIN INDEX
    sigmaR = 0
    for i in range(n):
        sigmaR += R(i, n, X, partition, reps) # Sum of max(Rij) = R
    return sigmaR / n                           # DB = Sum(R)/n 


def DBRepeat(n, X, t): ### GENERATE AND EVALUATE t NUMBER OF PARTITIONS GENERATE BY t USES OF THE KMEANS
    out = zeros(t)
    for i in range(t):
        partition, reps = kmeans(X, n, 100)
        out[i] = DB(n, X, partition, reps)
    return out

lim = 11    ### HOW MANY CLUSTERS TO TEST UP TO -1
repeat = 10 ### HOW MANY TIMES TO REPEAT
out = zeros((lim,2))
out[:,0] = range(lim)
for n in range(2,lim):
    out[n,1] = mean(DBRepeat(n, clusterVal, repeat))
    
out = out[2:]

plt.plot(out[:,0],out[:,1])
plt.xlabel("K Value")
plt.ylabel("Davies-Bouldin Value")
plt.title("Average Davies-Bouldin Value for different K Values over",repeat,"trials")
````
