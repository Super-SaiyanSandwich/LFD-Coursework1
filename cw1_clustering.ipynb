- **Euclidean Distance**

The simplest method of measuring the distance from a point to another. Used in testing the KMeans function before adding other methods of measuring distance, however is left for comparison.

````python
### EUCLIDEAN DISTANCE ###
def d(point, centroid, data, partition):
    out = 0
    for i in range(point.shape[0]):
        out += (point[i] - centroid[i])**2
    return sqrt(out)


### EUCLIDEAN DISTANCE INITIAL CLUSTER. FUNCTION REQUIRED TO BE SET AS MAY CHANGE WITH THE MAHALANOBIS FUNCTIONS ###
def initialCluster(data, reps, K):
    print("EUCLIDEAN DISTANCE")
    
    n = data.shape[0]
    partition = zeros((K,n))
    for i in range(n):        
        point = data[i]
        distances = zeros(K)
        for j in range(K):
            out = 0
            for k in range(point.shape[0]):
                out += (point[k] - reps[j][k])**2
            distances[j] = sqrt(out)
        partition[argmin(array(distances))][i] = 1
        
    for i in range(K):
            t = partition[i] == 1
            repData = data[t]
            reps[i] = mean(repData, axis=0,dtype=np.float32)
    return partition
````

- **Mahalanobis Distance**

As required by the coursework definition, the Mahalanobis function. As the Mahalanobis Distance uses a set of data to form the covariance, there are two different datasets that could be used. The first is all the data from all the clusters, and the second is just the data from the cluster being updated. As the calculation of the distance requires a set of data, on the initial pass, if only data from the cluster is being used, it will not work. For this reason, the "initial cluster" must be calculated using euclidean distance or using all the data. As is was unsure which would be the "best" method of implementing Mahalanobis below there is a series of variant methods based on Mahalanobis distance:
  * Using all the data, with Euclidean initial
  * Using all the data, with Mahalanobis initial
  * Using just the cluster data, with Euclidean initial
  * Using just the cluster data, with Mahalanobis initial
  
##### All data, Euclidean initial evaluation:

````python
### MAHALANOBIS DISTANCE: ALL DATA, EUCLIDEAN INITIAL ###
def d(point, centroid, data, partition):
    delta = point - centroid
    return sqrt( dot( dot( delta.T , linalg.inv(cov(data.T))), delta))

def initialCluster(data, reps, K):
    print("MAHALANOBIS DISTANCE: ALL DATA, EUCLIDEAN INITIAL") ## FOR IDENTIFYING THE SETTINGS WHEN RUNNING
    
    n = data.shape[0]
    partition = zeros((K,n))
    for i in range(n):        
        point = data[i]
        distances = zeros(K)
        for j in range(K):
            out = 0
            for k in range(point.shape[0]):
                out += (point[k] - reps[j][k])**2
            distances[j] = sqrt(out)
        partition[argmin(array(distances))][i] = 1
        
    for i in range(K):
            t = partition[i] == 1
            repData = data[t]
            reps[i] = mean(repData, axis=0,dtype=np.float32)
    return partition
````

- **Around a Circle**

As the data is always standarized, this function pick points based on evenly space points around a circle, radius 1. Works far more consistently as it avoids the issue of choosing several points too close within one cluster. Also helps compare the two distance functions by always starting with the same points. However it loses the randomness of the intial clusterings meaning the result of the kMeans algorithm is always the same.

````python
### DEFINED POINTS AROUND A CIRCLE
def initialReps(data, k):
    print("POINTS AROUND A CIRCLE INITIAL REPRESENTATIVES")
    d = data.shape[1]

    out = zeros((k,d))
    
    for i in range(k):        
        it = (2*pi*i)/k
        for j in range(d - 1):
            out[i,j] = sin(it)**j * cos(it)
        out[i,d-1] = sin(it)**(d-1)
    return out
````

- **K-means++**
An algorithm suggested in 2007 by David Arthur and Sergei Vassilvitskii, it is a way to avoid the sometimes poor initial clustering found by the other two methods and aims to seed the intial clusters in a particular way. The algorithms follows:
    1. Choose one center with a uniform randomness from among all data points.
    2. For each data point x, compute D(x), the distnace between x and the nearest center that has already been chosen
    3. Choose one new data point at random as a new center, using a weighted probability proportional to D(x)^2
    4. Repeat steps 2 and 3 until k centers have been chosen
    
This method takes far longer than the previous, however this seeding yields considerable improvement in the final error of k-means.

````python
### K-MEANS++ ###
import random

def dist(point, centroid):
    out = 0
    for i in range(point.shape[0]):
        out += (point[i] - centroid[i])**2
    return sqrt(out)

def initialReps(data, k):
    print("K-MEANS++ INITIAL REPRESENTATIVES")
    dim = data.shape[1]
    n = data.shape[0]
    
    out = zeros((k,dim))
    
    out[0] = random.choice(data)
    part = ones(n)
    
    for i in range(1,k):
        dMatrix = full(n, inf)
        for j in range(n):
            for k in range(i):
                dMatrix[j] = min(dMatrix[j], dist(data[j], out[k])**4)
        total = sum(dMatrix)
        r = random.uniform(0,total)
        for point in range(n):
            r -= dMatrix[point]
            if r <= 0:
                out[i] = data[point]
                break
        
        
    return out
        
        
````

#### Miscellaneous Functions:

Functions that are used in the project, but not explicitly part of the kMeans algorithm.

````python
### Takes a standard (n.k) binary matrix partition and converts it to a n.1 matrix ###
def flatten(partition):
    out = zeros(partition.shape[1])
    for i in range(1,partition.shape[0]):
        out += i*partition[i]
    return out
````

````python
### Takes a (n.1) matrix and converts it to a standard n.k partition matrix ###
def expand(flat):
    out = zeros((max(flat) + 1,flat.shape[0]))
    for i in range(flat.shape[0]):
        out[flat[i]][i] = 1
    return out
````

````python
### A way of comparing the result to the targets ###
def errors(partition, target):
    target = expand(target)
    
    partition.sort()
    target.sort()
    
    partition = flatten(partition)
    target = flatten(target)
    
    e = 0
    
    for i in range(target.shape[0]):
        if target[i] != partition[i]:
            e += 1
    
    return e    
````
